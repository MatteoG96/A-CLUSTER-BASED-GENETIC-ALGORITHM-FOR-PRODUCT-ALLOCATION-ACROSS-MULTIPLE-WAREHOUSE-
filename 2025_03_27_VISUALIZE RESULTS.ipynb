```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter

# ——— Your palette mapping ———
palette = {
    "NO CLUSTER": "#000000",
    "CLUSTER 1":  "#aec7e8",
    "CLUSTER 2":  "#1f77b4",
    "CLUSTER 3":  "#084c8d",
    "CLUSTER 4":  "#d0e4f7",
    "CATELAN GO": "green",
    "CATELAN GP": "orange",
    "CATELAN GS": "red",
    "CATELAN BS": "magenta",
}

# ——— Define driver groups ———
cluster_drivers = ["NO CLUSTER", "CLUSTER 1", "CLUSTER 2", "CLUSTER 3", "CLUSTER 4"]
catelan_group   = ["CLUSTER 1", "CATELAN GO", "CATELAN GP", "CATELAN GS", "CATELAN BS"]

# ——— Compute sort orders for cluster group ———
order_cluster_routes = (
    df_all[df_all["DRIVER"].isin(cluster_drivers)]
    .sort_values("% Rotte completate in AB", ascending=True)
    ["DRIVER"]
    .tolist()
)
order_cluster_orders = (
    df_all[df_all["DRIVER"].isin(cluster_drivers)]
    .sort_values("% Ordini completati in AB", ascending=True)
    ["DRIVER"]
    .tolist()
)

# ——— Compute sort orders for Catelan vs Cluster1 group ———
order_catelan_routes = (
    df_all[df_all["DRIVER"].isin(catelan_group)]
    .sort_values("% Rotte completate in AB", ascending=True)
    ["DRIVER"]
    .tolist()
)
order_catelan_orders = (
    df_all[df_all["DRIVER"].isin(catelan_group)]
    .sort_values("% Ordini completati in AB", ascending=True)
    ["DRIVER"]
    .tolist()
)

# ——— Figure 1: Compare all cluster approaches ———
fig, axes = plt.subplots(ncols=2, figsize=(16, 6), sharey=True)

# % Rotte Completate in AB for clusters
sns.barplot(
    data=df_all[df_all["DRIVER"].isin(cluster_drivers)],
    x="DRIVER",
    y="% Rotte completate in AB",
    order=order_cluster_routes,
    palette=palette,
    dodge=False,
    ax=axes[0]
)
axes[0].set_title("% Rotte Completate in AB (Clusters)")
axes[0].set_ylim(0, 100)
axes[0].set_ylabel("Percentuale")
axes[0].yaxis.set_major_formatter(PercentFormatter())
axes[0].tick_params(axis="x", rotation=20)

# % Ordini Completati in AB for clusters
sns.barplot(
    data=df_all[df_all["DRIVER"].isin(cluster_drivers)],
    x="DRIVER",
    y="% Ordini completati in AB",
    order=order_cluster_orders,
    palette=palette,
    dodge=False,
    ax=axes[1]
)
axes[1].set_title("% Ordini Completati in AB (Clusters)")
axes[1].set_ylim(0, 100)
axes[1].set_ylabel("")  # shared
axes[1].yaxis.set_major_formatter(PercentFormatter())
axes[1].tick_params(axis="x", rotation=20)

plt.tight_layout()
plt.show()


# ——— Figure 2: Compare Cluster 1 vs Catelan approaches ———
fig, axes = plt.subplots(ncols=2, figsize=(16, 6), sharey=True)

# % Rotte Completate in AB for Cluster1 vs Catelan
sns.barplot(
    data=df_all[df_all["DRIVER"].isin(catelan_group)],
    x="DRIVER",
    y="% Rotte completate in AB",
    order=order_catelan_routes,
    palette=palette,
    dodge=False,
    ax=axes[0]
)
axes[0].set_title("% Rotte Completate in AB (Cluster1 vs Catelan)")
axes[0].set_ylim(0, 100)
axes[0].set_ylabel("Percentuale")
axes[0].yaxis.set_major_formatter(PercentFormatter())
axes[0].tick_params(axis="x", rotation=20)

# % Ordini Completati in AB for Cluster1 vs Catelan
sns.barplot(
    data=df_all[df_all["DRIVER"].isin(catelan_group)],
    x="DRIVER",
    y="% Ordini completati in AB",
    order=order_catelan_orders,
    palette=palette,
    dodge=False,
    ax=axes[1]
)
axes[1].set_title("% Ordini Completati in AB (Cluster1 vs Catelan)")
axes[1].set_ylim(0, 100)
axes[1].set_ylabel("")  # shared
axes[1].yaxis.set_major_formatter(PercentFormatter())
axes[1].tick_params(axis="x", rotation=20)

plt.tight_layout()
plt.show()
```
