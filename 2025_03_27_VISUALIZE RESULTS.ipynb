## NUMBER OF CODE & STOCK

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ——— Melt for codes and stock ———
df_codes_melted = df_all.melt(
    id_vars=["DRIVER"],
    value_vars=["Codici in A", "Codici in B"],
    var_name="Categoria",
    value_name="Valore"
)

df_stock_melted = df_all.melt(
    id_vars=["DRIVER"],
    value_vars=["Stock [m3] in A", "Stock [m3] in B"],
    var_name="Categoria",
    value_name="Valore"
)

# ——— Palette definition ———
palette = {
    "NO CLUSTER": "#000000",
    "CLUSTER 1":    "#aec7e8",
    "CLUSTER 2":    "#1f77b4",
    "CLUSTER 3":    "#084c8d",
    "CLUSTER 4":    "#d0e4f7",
    "CATELAN GO":    "green",
    "CATELAN GP":    "orange",
    "CATELAN GS":    "red",
    "CATELAN BS":    "magenta",
}

# ——— Capacity thresholds ———
max_capacity_A   = 25000 * 0.37
max_capacity_B   = 6800  * 0.37
min_utilization_B = max_capacity_B * 0.5  # if you want to show this

# ——— Create figure with two axes ———
fig, axes = plt.subplots(ncols=2, figsize=(16, 6), sharey=False)

# Left subplot: Number of stored SKU
sns.barplot(
    data=df_codes_melted,
    x="Categoria",
    y="Valore",
    hue="DRIVER",
    palette=palette,
    ax=axes[0]
)
axes[0].set_ylabel("Number of stored SKU")
axes[0].set_xlabel("")
axes[0].legend(title="APPROACH")

# Right subplot: Stock [m³] with capacity lines
sns.barplot(
    data=df_stock_melted,
    x="Categoria",
    y="Valore",
    hue="DRIVER",
    palette=palette,
    ax=axes[1]
)
# horizontal capacity lines
axes[1].axhline(y=max_capacity_A,   linestyle='--', linewidth=1.5, label='Warehouse A Max')
axes[1].axhline(y=max_capacity_B,   linestyle='--', linewidth=1.5, label='Warehouse B Max')
#axes[1].axhline(y=min_utilization_B, linestyle='--', linewidth=1.5, label='Warehouse B Min Util')
axes[1].set_ylabel("Stock [m³]")
axes[1].set_xlabel("")
axes[1].tick_params(axis='x', rotation=20)
axes[1].legend(title="Legenda")

# ——— Final layout & save/show ———
plt.tight_layout()

# Optional: one shared legend instead of two
# handles, labels = axes[1].get_legend_handles_labels()
# fig.legend(handles, labels, title="DRIVER/APPROACH", loc='upper center', ncol=len(handles))
# axes[0].get_legend().remove()
# axes[1].get_legend().remove()

# To save:
# fig.savefig(r'C:\…\Analisi_Codici_e_Stock.png', dpi=300)

plt.show()
```

## LOGISTIC FLOW

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ——— Melt all four DataFrames to long format ———
df_routes_melted = df_all.melt(
    id_vars=["DRIVER"],
    value_vars=[
        "% Rotte completate in AB",
        #"% Rotte completate in A",
        #"% Rotte completate in B"
    ],
    var_name="Categoria",
    value_name="Valore"
)

df_orders_melted = df_all.melt(
    id_vars=["DRIVER"],
    value_vars=[
        "% Ordini completati in AB",
        #"% Ordini completati in A",
        #"% Ordini completati in B"
    ],
    var_name="Categoria",
    value_name="Valore"
)

df_vol_routes_melted = df_all.melt(
    id_vars=["DRIVER"],
    value_vars=[
        #"Vol[m3] Rotte completati in A",
        #"Vol[m3] Rotte completati in B",
        "Vol[m3] Rotte completati in AB",
        #"Vol[m3] Rotte completati in AB (A)",
        #"Vol[m3] Rotte completati in AB (B)"
    ],
    var_name="Categoria",
    value_name="Valore"
)

df_vol_orders_melted = df_all.melt(
    id_vars=["DRIVER"],
    value_vars=[
        #"Vol[m3] Ordini completati in A",
        #"Vol[m3] Ordini completati in B",
        "Vol[m3] Ordini completati in AB",
        #"Vol[m3] Ordini completati in AB (A)",
        #"Vol[m3] Ordini completati in AB (B)"
    ],
    var_name="Categoria",
    value_name="Valore"
)

# ——— Create 2×2 subplots, sharing y-axis by row ———
fig, axes = plt.subplots(
    nrows=2, ncols=2,
    figsize=(16, 12),
    sharey='row'   # first row share %, second row share volumes
)

# Top-left: % Rotte Completate
sns.barplot(
    data=df_routes_melted,
    x="Categoria", y="Valore", hue="DRIVER",
    palette=palette, ax=axes[0, 0]
)
axes[0, 0].set_title("% Rotte Completate")
axes[0, 0].set_xlabel("")
axes[0, 0].set_ylabel("%")
axes[0, 0].set_ylim(0, 100)
axes[0, 0].legend(title="DRIVER")
for lbl in axes[0, 0].get_xticklabels():
    lbl.set_rotation(15)

# Top-right: % Ordini Completati
sns.barplot(
    data=df_orders_melted,
    x="Categoria", y="Valore", hue="DRIVER",
    palette=palette, ax=axes[0, 1]
)
axes[0, 1].set_title("% Ordini Completati")
axes[0, 1].set_xlabel("")
axes[0, 1].set_ylabel("%")
axes[0, 1].set_ylim(0, 100)
axes[0, 1].legend(title="DRIVER")
for lbl in axes[0, 1].get_xticklabels():
    lbl.set_rotation(15)

# Bottom-left: Volume [m³] Rotte Completati
sns.barplot(
    data=df_vol_routes_melted,
    x="Categoria", y="Valore", hue="DRIVER",
    palette=palette, ax=axes[1, 0]
)
axes[1, 0].set_title("Volume [m³] Rotte Completati")
axes[1, 0].set_xlabel("")
axes[1, 0].set_ylabel("m³")
axes[1, 0].legend(title="DRIVER")
for lbl in axes[1, 0].get_xticklabels():
    lbl.set_rotation(20)

# Bottom-right: Volume [m³] Ordini Completati
sns.barplot(
    data=df_vol_orders_melted,
    x="Categoria", y="Valore", hue="DRIVER",
    palette=palette, ax=axes[1, 1]
)
axes[1, 1].set_title("Volume [m³] Ordini Completati")
axes[1, 1].set_xlabel("")
axes[1, 1].set_ylabel("m³")
axes[1, 1].legend(title="DRIVER")
for lbl in axes[1, 1].get_xticklabels():
    lbl.set_rotation(20)

# ——— Final layout & show ———
plt.tight_layout()
plt.show()
```
