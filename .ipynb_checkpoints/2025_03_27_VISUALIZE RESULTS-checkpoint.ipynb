## AREA LIBERATA

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# --- PRIMA ANALISI ---
# Filtro solo 'Stock [m3] in B'
df_b1 = df_stock_melted[df_stock_melted['Categoria'] == 'Stock [m3] in B'].copy()

# Calcolo i valori
df_b1['Spazio liberato [m3] in A'] = df_b1['Valore']
df_b1['Spazio rimanente [m3] in B'] = (6800 * 0.37 - df_b1['Valore'])

# Riformatto in formato long con label semplificati
df_new1 = pd.concat([
    df_b1[['DRIVER']].assign(Categoria='Spazio liberato in A', Valore=df_b1['Spazio liberato [m3] in A']),
    df_b1[['DRIVER']].assign(Categoria='Spazio rimanente in B', Valore=df_b1['Spazio rimanente [m3] in B'])
], ignore_index=True)

# --- SECONDA ANALISI ---
df_b2 = df_stock_melted[df_stock_melted['Categoria'] == 'Stock [m3] in B'].copy()

# Calcolo valori normalizzati
df_b2['Spazio liberato [m3] in A'] = df_b2['Valore'] / 0.37 / 10 /0.4
df_b2['Spazio rimanente [m3] in B'] = (6800 * 0.37 - df_b2['Valore']) / 0.37 / 7 / 0.4

# Riformatto in formato long con label semplificati
df_new2 = pd.concat([
    df_b2[['DRIVER']].assign(Categoria='Spazio liberato in A', Valore=df_b2['Spazio liberato [m3] in A']),
    df_b2[['DRIVER']].assign(Categoria='Spazio rimanente in B', Valore=df_b2['Spazio rimanente [m3] in B'])
], ignore_index=True)

# --- PLOT COMBINATO ---
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=False)

# Primo plot
sns.barplot(
    data=df_new1,
    x="Categoria",
    y="Valore",
    hue="DRIVER",
    ci=None,
    palette=palette,
    ax=axes[0]
)
axes[0].set_title("")
axes[0].set_ylabel("Volume per materiali [m3]")
axes[0].set_xlabel("")
axes[0].legend(title="Legenda")

# Secondo plot
sns.barplot(
    data=df_new2,
    x="Categoria",
    y="Valore",
    hue="DRIVER",
    ci=None,
    palette=palette,
    ax=axes[1]
)
axes[1].set_title("")
axes[1].set_ylabel("Spazio per scaffalature e corridoi (m2)")
axes[1].set_xlabel("")
axes[1].legend(title="Legenda")

plt.tight_layout()
plt.savefig(r"C:\Users\Matteo.Gabellini\OneDrive - Alma Mater Studiorum Università di Bologna\DOTTORATO\2.CONSULENZA\AZIENDE\MELONI\3.PPT\FIGURE\Analisi_Spazio_Liberato_Confronto.png")
plt.show()
```

## PERSONALE ACCETTAZIONE - SPEDIZIONE

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Melt the DataFrame to long format ---
df_melted = df_all[['DRIVER', 'Accettazione', 'Spedizione',
                    'Personale Accettazione A', 'Personale Accettazione B',
                    'Personale Spedizione A', 'Personale Spedizione B']].melt(
    id_vars=['DRIVER', 'Accettazione', 'Spedizione'],
    value_vars=[
        'Personale Accettazione A',
        'Personale Accettazione B',
        'Personale Spedizione A',
        'Personale Spedizione B'
    ],
    var_name='Categoria',
    value_name='Valore'
)

# --- Extract macro-categoria and sede from 'Categoria' ---
df_melted['Macro'] = df_melted['Categoria'].apply(lambda x: 'Accettazione' if 'Accettazione' in x else 'Spedizione')
df_melted['Sede'] = df_melted['Categoria'].apply(lambda x: 'A' if x.endswith('A') else 'B')

# --- Unisci Accettazione e Spedizione in una colonna unica ---
df_melted['Flusso'] = df_melted['Accettazione'] + " - " + df_melted['Spedizione']

# --- Sottrai 3 ai valori per le categorie specifiche ---
#mask = df_melted['Categoria'].isin(['Personale Accettazione A', 'Personale Spedizione A'])
#df_melted.loc[mask, 'Valore'] = df_melted.loc[mask, 'Valore'] - 3

# --- Plot ---
g = sns.catplot(
    data=df_melted,
    kind='bar',
    row='Sede',
    y='Valore',
    x='Macro',
    col='Flusso',
    hue='DRIVER',
    height=4,
    aspect=1,
    ci = False,
    palette=palette
)

g.set_axis_labels("Attività", "Numero operatori")
g.set_titles("Acc-Sped: {col_name} | Magazzino: {row_name}")
g._legend.set_title("DRIVER")

plt.tight_layout()
plt.show()
```
